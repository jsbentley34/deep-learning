```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('binary.csv')
```

```python
data.head()
```

### Data Preprocessing 

```python
df = pd.concat([data, pd.get_dummies(data['rank'], prefix='rank')],axis=1)
df = df.drop('rank',axis=1)
df.head()
```

```python
for field in ['gre', 'gpa']:
    mean, std = df[field].mean(), df[field].std()
    df.loc[:,field] = (df[field]-mean)/std
```

```python
df.head()
```

```python
sample = np.random.choice(df.index, size=int(len(data)*0.9), replace=False)
data, test_data = df.ix[sample], df.drop(sample)
```

```python
x_train, y_train = data.drop('admit',axis=1), data['admit']
x_test, y_test = test_data.drop('admit', axis=1), test_data['admit']
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

weights = np.random.normal(scale=1 / x_train.shape[1]**.5, size=x_train.shape[1])
learning_rate = 0.1
epochs = 10000
last_loss = None
c=0
for e in range(epochs):
    del_w = np.zeros(len(weights))
    for x, y in zip(x_train.values, y_train.values):
        a = sigmoid(np.dot(x, weights))
        error = y - a
        
        error_term = error * a * (1 - a)
        
        del_w += error_term * x
    weights += learning_rate*error_term/x_train.shape[0]
    
    if e%100 == 0:
        out = sigmoid(np.dot(x_train,weights))
        loss = np.mean((out-y_train)**2)
        if last_loss and last_loss<loss:
            print("WARNING - Loss Increasing..")
            print("Train Loss: {}".format(loss))
            c+=1
            if c==3:
                break
        else:
            print("Loss Decreasing..")
            print("Train Loss: {}".format(loss))
        last_loss = loss
```

```python
pred = sigmoid(np.dot(x_test,weights))
```

```python
pred = pred>0.5
```

```python
accuracy = np.mean(pred==y_test)
```

```python
print("Accuracy of the model is: ",accuracy*100)
```
